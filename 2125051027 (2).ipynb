Perform Multiple Transformations

a. Resize an image to 256x256 pixels.
b. Rotate by 180 degrees clockwise.
c. Apply a vertical flip.
d. Display all transformations in a single subplot.

```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
image_url = 'https://www.hindustantimes.com/ht-img/img/2023/06/14/550x309/demon_slayer_season_3_release_1681007034048_1686759381730.jpg'
```

```python
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
```

```python
image = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Image from URL")
plt.show()
```

```python
resized_image = cv2.resize(img_rgb, (256, 256))
plt.imshow(resized_image)
plt.axis('off')
plt.title("Resized")
plt.show()
```

```python
rotated_image = cv2.rotate(resized_image, cv2.ROTATE_180)
plt.imshow(rotated_image)
plt.axis('off')
plt.title("Rotated")
plt.show()
```

```python
flipped_image = cv2.flip(rotated_image, 0)
plt.imshow(flipped_image)
plt.axis('off')
plt.title("Flipped")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(resized_image)
plt.title("Resized")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(rotated_image)
plt.title("Rotated")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(flipped_image)
plt.title("Flipped")
plt.axis('off')

plt.tight_layout()
plt.show()
```
